```python
from glob import glob
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Input,Lambda,Dense,Flatten
from keras.models import Model
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import AveragePooling2D
from tensorflow.keras.layers import Dropout
from tensorflow.keras.layers import Flatten
```

```python
image_files = glob(train_path + '/*/*.jp*g')
valid_image_files = glob(test_path + '/*/*.jp*g')
```

```python
image_size=[224,224]
batch_size=12
epochs=40
folders = glob(train_path + '/*')
mobilenetv2=MobileNetV2(weights='imagenet',include_top=False,input_shape=image_size+[3])
```

```python
for layer in mobilenetv2.layers:
    layer.trainable=False
    
x=mobilenetv2.output
x=AveragePooling2D(pool_size=(7,7))(x)
x=Flatten(name='Flatten')(x)
x=Dense(128,activation='relu')(x)
x=Dropout(0.5)(x)
prediction=Dense(len(folders),activation='softmax')(x)
```

```python
model=Model(inputs=mobilenetv2.input,outputs=prediction)
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer="adam",
             metrics=['accuracy'])
```

```python
genarator=ImageDataGenerator( rotation_range=20,
                              width_shift_range=0.1,
                              height_shift_range=0.1,
                              shear_range=0.1,
                              zoom_range=0.2,
                              horizontal_flip=True,
                              vertical_flip=False,
                              preprocessing_function=preprocess_input)
```

```python
test_gen=genarator.flow_from_directory(test_path,target_size=image_size)
print(test_gen.class_indices)
```

```python
labels=[None]*len(test_gen.class_indices)
for k,v in test_gen.class_indices.items():
    labels[v]=k
```

```python
for x,y in test_gen:
    print("min:", x[0].min(), "max:", x[0].max())
    plt.title(labels[np.argmax(y[0])])
    plt.grid(False)
    plt.imshow(x[0])
    plt.show()
    break   
```

```python
train_genarator=genarator.flow_from_directory(train_path,
                                              target_size=image_size,
                                              shuffle=True,
                                              batch_size=batch_size)
```

```python
validation_genarator=genarator.flow_from_directory(test_path,
                                              target_size=image_size,
                                              shuffle=False,
                                              batch_size=batch_size)
```

```python
results=model.fit_generator(train_genarator,
                           validation_data=validation_genarator,
                           epochs=epochs,
                           steps_per_epoch=len(image_files)//batch_size,
                           validation_steps=len(valid_image_files)//batch_size)
```

```python
model.save('mobilenet_v2.model')
```

```python
from tensorflow.keras.models import model_from_json
```

```python
model_json=model.to_json()
with open("model.json","w") as json_file:
    json_file.write(model_json)
model.save_weights("model.h5")
```

```python
json_file=open("model.json","r")
loaded_model_json=json_file.read()
json_file.close()
loaded_model=model_from_json(loaded_model_json)
loaded_model.load_weights("model.h5")
```

```python
loaded_model.compile(loss="categorical_crossentropy",optimizer="adam",metrics=["accuracy"])
```

```python
loaded_model.evaluate_generator(validation_genarator)
```
